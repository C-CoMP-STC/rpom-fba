```python
from cobra.io import read_sbml_model

queen = read_sbml_model("Rpom_05__no_probs.xml")
problematic = read_sbml_model("Rpom_05__problematic.xml")
```

```python
rxn_only_in_a = set(rxn.id for rxn in queen.reactions) - set(rxn.id for rxn in problematic.reactions)
rxn_only_in_a
```

```python
for rxn_a in queen.reactions:
    rxn_b = problematic.reactions.get_by_id(rxn_a.id)

    if rxn_a.reaction != rxn_b.reaction:
        print(f"Reaction mismatch for {rxn_a.id}: {rxn_a.reaction} vs {rxn_b.reaction}")

    if rxn_a.lower_bound != rxn_b.lower_bound:
        print(f"Lower bound mismatch for {rxn_a.id}: {rxn_a.lower_bound} vs {rxn_b.lower_bound}") 
```

```python
with queen:
    queen.reactions.get_by_id("EX_glc").lower_bound = -10

    sol = queen.optimize()

print(sol)
```
